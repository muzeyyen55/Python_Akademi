```python
df["upper_title"] = df.original_title.str.upper()
df[["original_title", "upper_title"]]
```

```python
import string

def remove_punc(text):
    title = text.translate(str.maketrans('', '', string.punctuation))
    return title

df["remove_punc"] = df.original_title.apply(remove_punc)

df[["original_title", "upper_title", "remove_punc"]]
```
